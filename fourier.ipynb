# illustration de la decomposition en series de fourier

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Paramètres du signal
f0 = 50  # Fréquence fondamentale (50 Hz)
fs = 1000 # Fréquence d'échantillonnage (1000 points par seconde)
t = np.linspace(0, 0.1, fs) # On regarde le signal pendant 0.1 seconde

# 2. Création d'un signal "mélangé" (Somme de Fourier)
# On mélange une onde de 50Hz (Amplitude 1) et une de 120Hz (Amplitude 0.5)
signal = np.sin(2 * np.pi * f0 * t) + 0.5 * np.sin(2 * np.pi * 120 * t)

# 3. Calcul de la décomposition (FFT)
spectre = np.abs(np.fft.fft(signal))
frequences = np.fft.fftfreq(len(t), 1/fs)

# 4. Affichage
plt.figure(figsize=(12, 5))

# Graphique du signal temporel s(t)
plt.subplot(1, 2, 1)
plt.plot(t, signal)
plt.title("Signal s(t) mélangé")
plt.xlabel("Temps (s)")

# Graphique du spectre (Les coefficients Cn)
plt.subplot(1, 2, 2)
plt.plot(frequences[:fs//10], spectre[:fs//10]) # On zoome sur les fréquences positives
plt.title("Décomposition (Coefficients Cn)")
plt.xlabel("Fréquence (Hz)")

plt.tight_layout()
plt.show()
```
